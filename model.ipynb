```python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
dataset_folder = "datasets"

def find_first_image(directory):
    for root, _, files in os.walk(directory): 
        for file in files:
            if file.lower().endswith((".png", ".jpg", ".jpeg")): 
                return os.path.join(root, file) 
    return None

img_path = find_first_image(dataset_folder)

if img_path:
    img = Image.open(img_path)  

    plt.imshow(img)
    plt.axis("off")  
    plt.title(f"Displaying: {img_path}") 
    plt.show()
else:
    print("No images found in the datasets folder.")
```
